```python
import pandas as pd

# Load the dataset
# file_path = '/mnt/data/Mozilla.csv'
data = pd.read_csv('MozillaPre.csv')

# Display the first few rows of the dataset
data.head()
```

```python
assignee_experience = data['Assignee'].value_counts().to_dict()

# Map the experience back to the dataset
data['Experience'] = data['Assignee'].map(assignee_experience)

# Display the first few rows to verify the changes
data


```

```python

# # Calculate the experience for each assignee
# assignee_experience = data['Assignee'].value_counts().reset_index()
# assignee_experience.columns = ['Assignee', 'Experience']

# # Merge the experience data back to the original dataset
# data = pd.merge(data, assignee_experience, on='Assignee', how='left')
# Convert 'Opened' to datetime and calculate the age of the bug in days
data['Opened'] = pd.to_datetime(data['Opened'], errors='coerce')
data['Closed'] = pd.to_datetime(data['Closed'], errors='coerce')
data['Age'] = (data['Closed'] - data['Opened']).dt.days

# Fill missing ages with the max age for open bugs
data['Age'] = data['Age'].fillna((pd.to_datetime('today') - data['Opened']).dt.days)

# Display the first few rows of the dataset with the added age column
data.head()
# Step 6: Map text values to numeric
priority_mapping = {'P1': 1, 'P2': 2, 'P3': 3, 'P4': 4, 'P5': 5}
severity_mapping = {'critical': 1, 'S1': 2, 'S2': 3, 'S3': 4, 'S4': 5, 'minor': 6, 'normal': 7}

data['Priority'] = data['Priority'].map(priority_mapping)
data['Severity'] = data['Severity'].map(severity_mapping)
data.head()
# Ensure that all feature columns are numeric
data['Experience'] = pd.to_numeric(data['Experience'], errors='coerce')
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Priority'] = pd.to_numeric(data['Priority'], errors='coerce')
data['Severity'] = pd.to_numeric(data['Severity'], errors='coerce')
# Drop rows with any NaN values in feature columns
data.dropna(subset=['Priority', 'Severity', 'Experience', 'Age'])

# Debugging: Print the data to ensure it's not empty
print(f"Data shape after preprocessing: {data.shape}")
print(data.head())
```

```python
import ahpy

# Define the AHP comparison matrix
comparisons = {
    ('Priority', 'Severity'): 3,
    ('Priority', 'Experience'): 5,
    ('Priority', 'Age'): 3,
    ('Severity', 'Experience'): 5,
    ('Severity', 'Age'): 4,
    ('Experience', 'Age'): 4
}

# Calculate AHP weights
# Step 3: Calculate AHP weights
criteria_ahp = ahpy.Compare(name='Criteria', comparisons=comparisons, precision=3, random_index='saaty')
criteria_weights = criteria_ahp.target_weights
print('AHP Criteria Weights:', criteria_weights)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Select features and target
features = ['Priority', 'Severity', 'Experience', 'Age']
target = 'Assignee Real Name'

# Check if the data is empty after preprocessing
if len(data) == 0:
    print("No data available after preprocessing.")
else:
    # Split the data
    X = data[features]
    y = data[target]

    # Debugging: Check the target variable
    print(f"Unique target values: {y.nunique()}")

    # Proceed with train-test split if data is not empty
    if len(data) > 0:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

        # Step 9: Apply AHP weights to the features
        for feature in features:
            X_train[feature] = X_train[feature] * criteria_weights[feature]
            X_test[feature] = X_test[feature] * criteria_weights[feature]

        # Step 10: Train the model
        model = RandomForestClassifier(n_estimators=300, random_state=1)
        model.fit(X_train, y_train)

        # Step 11: Predict on the weighted test set
        y_pred_weighted = model.predict(X_test)

        # Step 12: Evaluate the weighted model
        accuracy_weighted = accuracy_score(y_test, y_pred_weighted)
        precision_weighted = precision_score(y_test, y_pred_weighted, average='weighted')
        recall_weighted = recall_score(y_test, y_pred_weighted, average='weighted')
        f1_weighted = f1_score(y_test, y_pred_weighted, average='weighted')

        # Step 13: Print the evaluation results
        print(f'Weighted Accuracy: {accuracy_weighted:.2f}')
        print(f'Weighted Precision: {precision_weighted:.2f}')
        print(f'Weighted Recall: {recall_weighted:.2f}')
        print(f'Weighted F1 Score: {f1_weighted:.2f}')
       # Step 14: Define a function to rank assignees for new bug data
        def rank_assignees(new_bug_data):
            # Convert new bug data to a DataFrame
            new_bug_df = pd.DataFrame([new_bug_data], columns=features)

            # Apply AHP weights to the new bug data
            for feature in features:
                new_bug_df[feature] = new_bug_df[feature] * criteria_weights[feature]

            # Predict probabilities for each assignee
            assignee_probs = model.predict_proba(new_bug_df)

            # Get the class names (assignees)
            assignees = model.classes_

            # Create a DataFrame with assignees and their probabilities
            assignee_ranking = pd.DataFrame({
                'Assignee': assignees,
                'Probability': assignee_probs[0]
            })

            # Sort by probability in descending order and add ranking
            assignee_ranking.sort_values(by='Probability', ascending=False, inplace=True)
            assignee_ranking['Rank'] = range(1, len(assignee_ranking) + 1)

            return assignee_ranking

        # Step 15: Rank assignees for a new bug
        new_bug_data = {
            'Priority': 5,        # Example values for a new bug
            'Severity': 4,
            'Product': 101,
            'Component': 202,
            'Experience': 1,
            'Age': 45
         }

        assignee_ranking = rank_assignees(new_bug_data)
        # print("Assignee Ranking for the New Bug:")
        # print(assignee_ranking)
     # Display the top 10 assignees and their ranks, without index and probability columns
        top_10_assignees = assignee_ranking.head(10)[['Assignee', 'Rank']].reset_index(drop=True)
        print("Top 10 Assignees for the New Bug:")
        print(top_10_assignees)
    else:
        print("No data available for training after preprocessing.")
```
